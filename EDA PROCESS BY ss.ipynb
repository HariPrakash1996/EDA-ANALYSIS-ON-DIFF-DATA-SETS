```python
def basic_check(__input__):
    table=BeautifulTable
    a=[]
    a=__input__.columns
    table.columns_headers=["features","feature_datatype","unique values","missing_values","missing_value_%"]
    for i in a:
        table.append_row([i,
                          __input__[i].dtype,
                          __input__[i].nunique(),
                          __input__[i].isna().sum(),
                          ((__input__[i].isna().sum()) / (__input__[i].shape[0])*100)])
        
        
    print(table)
```
